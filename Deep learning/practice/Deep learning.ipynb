```python
#knn_model.evaluate(image_test[image_test['label']=='dog'])
dog_image_train=image_train[image_train['label']=='dog']
cat_image_train=image_train[image_train['label']=='cat']
auto_image_train=image_train[image_train['label']=='automobile']
bird_image_train=image_train[image_train['label']=='bird']

dog_image_test=image_test[image_test['label']=='dog']
auto_image_test=image_test[image_test['label']=='automobile']
cat_image_test=image_test[image_test['label']=='cat']
bird_image_test=image_test[image_test['label']=='bird']
```

```python
dog_knn_model=gl.nearest_neighbors.create(dog_image_train,features=['deep_features'],label='id')
cat_knn_model=gl.nearest_neighbors.create(cat_image_train,features=['deep_features'],label='id')
auto_knn_model=gl.nearest_neighbors.create(auto_image_train,features=['deep_features'],label='id')
bird_knn_model=gl.nearest_neighbors.create(bird_image_train,features=['deep_features'],label='id')
```

```python
dog_cat_neighbors=cat_knn_model.query(cat_image_test,k=1)
dog_auto_neighbors=auto_knn_model.query(auto_image_test,k=1)
dog_bird_neighbors=bird_knn_model.query(bird_image_test,k=1)
dog_dog_neighbors=dog_knn_model.query(dog_image_test,k=1)
```

```python
dog_distances=gl.SFrame({'dog_dog':dog_dog_neighbors['distance'],'dog_cat':dog_cat_neighbors['distance'],'dog_auto':dog_auto_neighbors['distance'],'dog_bird':dog_bird_neighbors['distance']})
```

```python
dog_distances.head()
```

```python
def is_dog_correct_row(row):
    if (row['dog_auto']>row['dog_dog'] and row['dog_bird']>row['dog_dog'] and row['dog_cat']>row['dog_dog']):
        return 1
    else:
        return 0
    
```

```python
dog_distances.apply(is_dog_correct_row)
```

```python
dog_distances.apply(is_dog_correct_row).sum()
```
